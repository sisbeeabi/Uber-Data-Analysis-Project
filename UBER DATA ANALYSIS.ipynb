```python
#conversion

df.pickup_dt = pd.to_datetime(df.pickup_dt)
                            
# extracting month, day, time etc
df["day_name"] = df.pickup_dt.dt.day_name()
df["month_name"] = df.pickup_dt.dt.month_name() 
df["start_hour"] = df.pickup_dt.dt.hour
df["day"] = df.pickup_dt.dt.day
df["year"] = df.pickup_dt.dt.year                           
```

```python
# Recommendations
1. It was observed that weather variables have negligible correllation with pickups and do not affect pickups
2. In planning to meet demands of customers for pickups, weather forecasts should be disregarded
3. Over 10% of data on borough was missing. This could render the count of borough unreliable and distort demand patterns.
   The company should endeavour to have all data on borough recorded.
4. The total  number of pickups on non holidays far exceed those on holidays.
   This could be because people have to move around a lot on non holidays eg. to the office, 
   and prefer to rest at home on holidays. This is the most significant factor affecting pickups.
   The company should make adequate resources available on non holid  ays to accommodate large pickups.
5. Days of the week also affect total pickups. Pickups are highest on Saturdays and lowest on Mondays. 
   The company needs to note this and make adequate arrangements.
6. Total pickups are also affected by the hour of the day. Total pickups are highest about 8.00 pm daily   
   with a lower but significant peak at about 10.00am.This needs to be noted and arranged for.
7. Pickups were also noted to be highest in June
```
